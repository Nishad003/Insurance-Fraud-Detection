```python
valid_cities = [
    "Delhi","Ahmedabad","Mumbai","Bengaluru",
    "Jaipur","Pune","Hyderabad","Kolkata","Chennai"
]

data["city"] = data["city"].str.strip().str.title()
data.loc[~data["city"].isin(valid_cities), "city"] = "Other"
```

```python
data['policy_start_date'] = pd.to_datetime(
    data['policy_start_date'], 
    errors= 'coerce', 
    dayfirst=True
)
```

```python
data['claim_date'] = pd.to_datetime(
    data['claim_date'], 
    errors= 'coerce', 
    dayfirst=True
)
```

```python
data["policy_start_date"] = data["policy_start_date"].fillna(
    data["policy_start_date"].median()
)
```

```python
data["claim_date"] = data["claim_date"].fillna(
    data["claim_date"].median()
)
```

```python
print(data['policy_start_date'].isnull().sum())
print(data['claim_date'].isnull().sum())
```

```python
colmuns = ['customer_age', 'annual_premium',
       'claim_amount', 'past_claims_count', 'days_since_policy_start',
       'documents_submitted']


for col in colmuns:
    count = 0
    for i in data[col]:
        if i < 0:
            count += 1
    print(f"{col} has {count} Negative Values")
    print()
print()
```

```python
for col in colmuns:
    data.loc[data[col] < 0, col] = np.nan
```

```python
print(data.isnull().sum())
```

```python
fill_strategy = {
    "customer_age": data["customer_age"].median(),
    "annual_premium": data["annual_premium"].median(),
    "claim_amount": data["claim_amount"].median(),
    "past_claims_count": data["past_claims_count"].median(),
    "days_since_policy_start": data["days_since_policy_start"].median(),
    "documents_submitted": data["documents_submitted"].median(),
}

data = data.fillna(fill_strategy)
```

```python
print(data.isnull().sum())
```

```python
num_cols = data.select_dtypes(include="number").columns

for col in num_cols:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    outliers = ((data[col] < q1 - 1.5*iqr) | (data[col] > q3 + 1.5*iqr)).sum()
    print(col, outliers)
```

```python
# Claim before policy start = suspicious
(data["claim_date"] < data["policy_start_date"]).sum()

```

```python
for col in data.select_dtypes(include="object"):
    print(col, data[col].nunique())
```

```python
num_cols = data.select_dtypes(include="number").columns

for col in num_cols:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    
    outliers = data[(data[col] < lower) | (data[col] > upper)][col]
    
    print(f"\nOutliers in {col}:")
    print(outliers)
```

```python
mask = data["claim_date"] < data["policy_start_date"]

# swap values when reversed
data.loc[mask, ["policy_start_date", "claim_date"]] = \
data.loc[mask, ["claim_date", "policy_start_date"]].values
```
